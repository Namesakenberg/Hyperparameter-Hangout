```python
import numpy as np

# create a column to give price
coffee["price"]=3.50
# the same price is given to both the latte and espresso

# make new column to give price of 3.50 to latte else 4
coffee["new_price"]=np.where(coffee["Coffee Type"]=="Latte",5.99,3.99)
```

```python
coffee.head()
```

```python
# remove the older price column
coffee=coffee.drop(columns=['price'])
```

```python
# add a new revenue column
coffee["revenue"]=coffee['Units Sold']*coffee['new_price']
```

```python
coffee.rename(columns={'new_price':'price'})
# rename the column 'new price' to 'price'
```

```python
# handling the null values
# to handle , adding some null values to the data
coffee.loc[[0,1],'Units Sold']=np.nan
```

```python
coffee
# now the first 2 values are NaN
```

```python
# now to replace the values we can use
coffee.fillna(coffee['Units Sold'].mean())
# replaced the nan walves with the mean of the
```

```python
# another way is to use the dropna but doing that will drop the entire row in which the null values are there
coffee.dropna()
```

```python
coffee.info()
```

```python
# advanced pandas functions
# .shift() , .rank() , .rolling() , .cumsum()
```

```python
coffee['yesterday_revenue']=coffee['revenue'].shift(1)
coffee
```

```python
coffee['percent_change']=coffee['revenue']/coffee['yesterday_revenue']
coffee
```

Filtering on the basis of height and weight

```python
# find the athletes taller than 200cm
bios.loc[(bios['height_cm']>200) & (bios['born_country']=='USA')]
```

```python
# make a column to store the first name of the athlete
# from the full name as it has spaces split it and retrun the part before space
bios['fname']=bios['name'].str.split(' ').str[0]
```

```python
import pandas as pd

# Convert 'born_date' column to datetime format
bios['born_date'] = pd.to_datetime(bios['born_date'])

# Extract the year and store it in 'born_year'
bios['born_year'] = bios['born_date'].dt.year

```

```python
bios.head()
```

```python
bios.info()
```

```python
# classify the athletes int short , avg and long on the basis of their height
bios['height_category']=bios['height_cm'].apply(lambda x:'short' if x<165 else ('avg' if x<185 else'Tall'))
```

```python
bios[bios['born_country'].isin(["USA","GBR","FRA"]) & (bios['name'].str.startswith("Keith"))]
#find people from countries gbr , fra and usa having name starting with 'Keith
```

```python
# only print the players with small height
bios[bios['height_category']=='short']
```

```python
bios.query('born_country=="USA" and born_city=="Seattle"')
# use the query function to find the people living in a specific city of a speciific country
```
