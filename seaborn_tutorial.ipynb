```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/housing.csv')
```

```python
df.head()
```

```python
#relational plot in seaborn
g=sns.relplot(kind='scatter',data=df,x='longitude',y='latitude')
g.set(title="geographical location of areas in California")
# data  = refers to the dataframe
# enter x and y axis
```

```python
g=sns.relplot(kind="scatter",data=df,x='longitude',y='latitude',hue='median_house_value')
g.set(title="geographical location of areas in California")

# hue : Maps data points to different colors based on a categorical or numerical variable.
# hue here is done on a numerical value
```

```python
g=sns.relplot(kind="scatter",data=df,x='longitude',y='latitude',hue='ocean_proximity')
g.set(title="geographical location of areas in California")
# hue on a categorical data
```

```python
# make a new column which would indicate whether the median proce of a certain house is greater than the median of the median house prices
df['above_median_house_value']= (df.median_house_value > df.median_house_value.median()).astype(int)
```

```python
df
# see the additional column at the end
```

```python
# create an hue for the binary variable
sns.relplot(kind='scatter',data=df,x='longitude',y='latitude',hue='above_median_house_value')
```

```python
# make a categorical plot on the total rooms and the ocean proximity column

sns.catplot(data=df,x='total_rooms',y="ocean_proximity")
```

```python
sns.catplot(data=df,x='median_house_value',y="ocean_proximity",hue='above_median_house_value')
```

```python
sns.boxplot(x='median_house_value',y='ocean_proximity',data=df)
```

```python
sns.boxplot(x='median_house_value',y='ocean_proximity',data=df,hue='above_median_house_value')
```

```python
from tensorflow.keras.utils import get_file
from copy import deepcopy
import os

zip_path = get_file(
    origin='https://storage.googleapis.com/tensorflow/tf-keras-datasets/jena_climate_2009_2016.csv.zip',
    fname='jena_climate_2009_2016.csv.zip',
    extract=True)

csv_path, _ = os.path.splitext(zip_path)

weather_df = pd.read_csv(csv_path)

weather_df
```

```python
weather_df.info()
# from the data types given below we can see that the 1st column (Date Time) is not of datetime type instead it isof type string
```

```python
weather_df.shape
```

```python
weather_df = weather_df[::6*24][:365]
weather_df.reset_index(drop=True, inplace=True)

weather_df_full = deepcopy(weather_df)

weather_df = weather_df[['Date Time', 'p (mbar)', 'T (degC)']]

weather_df
```

```python
weather_df.info()
```

```python
weather_df['Date Time']
```

```python
# convert this column from object to date time type
weather_df['Date Time'] = pd.to_datetime(weather_df['Date Time'],format='%d.%m.%Y %H:%M:%S')
```

```python
weather_df['Date Time']
# now the column is into proper format
# when the plotting is to be done and the date and time are the inputs then ensure that they are in datetime datatype
```

```python
weather_df.head()
```

```python
sns.lineplot(data=weather_df,x='Date Time',y='T (degC)')
```

```python
# make a new column to find if the pressure of air at the particular time is greater than to the mean of the pressure
weather_df['pressure_compwmean']=(weather_df['p (mbar)']	> weather_df['p (mbar)'].mean()).astype(int)
```

```python
weather_df
```

```python
weather_df['pressure_compwmean'].value_counts()
# 1 = pressure higher than mean
# 2 = pressure lower than mean
```

```python
sns.lineplot(data=weather_df,x='Date Time',y='T (degC)',hue='pressure_compwmean')
```
