Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
balance_data = pd.read_csv('/content/balance-scale.data')
```

```python
balance_data.head()
```

```python
balance_data.shape
```

Train test split on the data

```python
from sklearn.model_selection import train_test_split
X=balance_data.values[:,1:5]
y=balance_data.values[:,0]
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=100)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_gini=DecisionTreeClassifier(criterion="gini",random_state=100,max_depth=3,min_samples_leaf=5)
clf_gini.fit(X_train,y_train)
```

```python
clf_entropy=DecisionTreeClassifier(criterion="entropy",random_state=100,max_depth=3,min_samples_leaf=5)
clf_entropy.fit(X_train,y_train)
```

classification by the gini index

```python
y_pred_gini=clf_gini.predict(X_test)
print("predicted values :")
print(y_pred_gini)
```

classification by entropy

```python
y_pred_entropy=clf_entropy.predict(X_test)
print("predicted values :")
print(y_pred_entropy)
```

actual values

```python
print(y_test)
```

confusion matrix

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred_gini)
```

```python
confusion_matrix(y_test,y_pred_entropy)
```

accuracy score

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_gini)*100
```

```python
accuracy_score(y_test,y_pred_entropy)*100
```

Calssification report

```python
from sklearn.metrics import classification_report
classification_report(y_test,y_pred_gini,zero_division=0)
```

```python
classification_report(y_test,y_pred_entropy,zero_division=0)
```

printing the classification tree

```python
from sklearn.tree import plot_tree
def plot_decision_tree(clf_object,feature_names,class_names):
  plt.figure(figsize=(15,10))
  plot_tree(clf_object,filled=True,
            feature_names=feature_names,class_names=class_names,rounded=True)
  plt.show()
```

```python
# Visualizing the Decision Trees
plot_decision_tree(clf_gini, ['X1', 'X2', 'X3', 'X4'], ['L', 'B', 'R'])
plot_decision_tree(clf_entropy, ['X1', 'X2', 'X3', 'X4'], ['L', 'B', 'R'])
```
