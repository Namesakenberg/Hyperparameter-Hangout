# histograms

```python
# create a normal distribution curve
ages=np.random.normal(20,1.5,1000)
```

```python
#  working behind np.random.normal(20,1.5,1000)
#  np.random.normal(20,1.5,1000)
# The function generates random numbers that follow a normal (Gaussian) distribution with the specified mean and standard deviation
# 20 = mean : most of the values will be around mean
# 1.5 = spreading about mean , if std dev is more then it will go more or lesser than given mean
# 1000 = size
```

```python
plt.hist(ages)
```

```python
# using "bins"
# "bins" means the number of coulmns in the histogram
```

```python
# setting bins to 3
# 14-18
# 18-21
# 22-25
```

```python
plt.hist(ages,bins=[ages.min(),18,21,ages.max()])
```

```python
plt.hist(ages,bins=20,cumulative=True)
```

#plot customization

```python
# sample data :
years =[2014,2015,2016,2017,2018,2019,2020,2021]
income =[55,56,62,61,72,72,73,75]
income_ticks=list(range(50,81,2))
```

```python
plt.plot(years,income)
plt.show()
```

```python
# give the labels to the axes
plt.plot(years,income)
plt.title("yearly income of john in USD")
plt.xlabel("year")
plt.ylabel("income")
plt.yticks(income_ticks,[f"${x}k"for x in income_ticks])
plt.show()
```

# legends and multiple plots

```python
# different representations on the same diagram to represent different things is a legend
```

```python
# get some data
stock_a=[100,102,99,101,101,100,102]
stock_b=[90,95,102,104,105,103,109]
stock_c=[110,115,100,105,100,98,95]
# plot the data with labels
plt.plot(stock_a,label="company1")
plt.plot(stock_b,label="company2")
plt.plot(stock_c,label="company3")
# name the axis and the entrie plot
plt.title("stock price during 6 days")
plt.xlabel("day")
plt.ylabel("stock price")
# legend = place to represent the scale on the graph
plt.legend(loc="upper right")
#show
plt.show()
```
