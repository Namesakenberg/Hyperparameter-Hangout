#parametic density estimation

```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import normal
sample = normal(loc =10,scale = 5,size=1000) # data generated froma normal function
```

```python
sample
```

```python
sample.mean()
```

```python
# make a histogram to understand what kind of data we have
```

```python
plt.hist(sample,bins=50)
```

```python
# calculating mean and standard deviation from the data
sample_mean = sample.mean()
sample_std=sample.std()
```

from the graph drawn above we can see that the graph is somewhat like normal distribution

image.png

this is the formula for PDF(probability density function) of normal distribution

in this equation we

substitute the mu with sample mean

and

substitute the sigma with standard deviation

and provide x to this equation to get the PDF for vaalue of x

```python
# get the equation
from scipy.stats import norm
dist = norm(sample_mean,sample_std)
```

find minimum and maximum value in the data , from the range of the data  (max-min) we get 100 points randomly using np.linspace and put every point into the equation in place of x

```python
sample .min()
```

```python
sample.max()
```

```python
values = np.linspace(sample.min(),sample.max())
```

```python
probability_density = [dist.pdf(value) for value in values]
# pass all the values in equation and append the probability density in the list
```

```python
# plot the graph of the output points
```

```python
plt.hist(sample , bins=50,density=True )
plt.plot(values,probability_density)
```

#KDE

```python
# generate a sample
sample1 = normal(loc=20 , scale =5 , size=300)
sample2= normal(loc=40 , scale =5 , size=700)
sample= np.hstack((sample1 ,sample2))

# generate normal distribution data of 1000 points
# loc = mean
# scale = std dev
# hstack = horizontally concat sample 1 and sample 2

sample
```

```python
plt.hist(sample, bins = 50)
```

#### the graph above does not follow any sort of known dstribution and we have to find the pdf for this graph

```python
# hence as the graph is not known before we will use the non parametric distribution using KDE
# KDE = gaussian as it is mostly used
from sklearn.neighbors import KernelDensity
model = KernelDensity(bandwidth = 0.5 , kernel ='gaussian')
# convert data to 2D arrray as most ml algorithms work on 2d data
sample = sample .reshape((len(sample),1))
# now sample = (1000,1)
model.fit(sample)
```

```python
values=np.linspace(sample.min(),sample.max(),100)
values = values.reshape((len(values),1))
```

```python
pdf_kde = model.score_samples(values)
# score_samples is used to get the density estimate of the samples
# it is a pdf in logarithmic form as it is easy to represent smaller values in log form
# take exponential to convert logarithmic values to normal form
pdf_kde = np.exp(pdf_kde)
```

```python
# plot graph
plt.hist(sample,bins =50 ,density =True)
plt.plot(values,pdf_kde)
```

```python
# lower the bandwidth peaked is the curve
# higher is the bandwidth smoother is the curve
```
